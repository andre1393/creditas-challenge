# Utilization and claims dataset

```python
utilization_and_claims = pd.read_csv('../../datasets/DCA_Dataset_Utilization_and_Claims.csv', encoding = 'ISO-8859-1')
```

```python
utilization_and_claims['Claims'] = utilization_and_claims['Claims'].apply(lambda x: float(x.replace('$', '').replace(',','')))
```

```python
utilization_and_claims['Cumulative Loans'] = utilization_and_claims['Cumulative Loans'].apply(lambda x: np.nan if x is np.nan else x.replace(',',''))
```

# Human development index dataset

```python
hdi = pd.read_csv('../../datasets/hdi_human_development_index_hdig_value.csv')
```

```python
hdi.head()
```

```python
hdi.describe()
```

```python
hdi.info()
```

```python
fig, ax = plt.subplots(figsize = (10, 9))
ax.plot('year', 'value', data = hdi[hdi['country_code'] == 'NOR'], label = 'Noruega')
ax.plot('year', 'value', data = hdi[hdi['country_code'] == 'USA'], label = 'Estados Unidos')
ax.plot('year', 'value', data = hdi[hdi['country_code'] == 'BRA'], label = 'Brasil')
ax.plot('year', 'value', data = hdi[hdi['country_code'] == 'NER'], label = 'Nigeria')
ax.grid()
ax.set_title('IDH no brasil de 1980 a 2013')
ax.set_xlabel('ano')
ax.set_ylabel('IDH')
ax.legend()
```
